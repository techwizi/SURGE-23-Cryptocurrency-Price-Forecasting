```python
import pandas as pd
from datetime import datetime
```

```python
data = pd.read_csv('coin_Bitcoin.csv')
data
```

```python
dates = []
for date in data['Date']:
  dates.append(pd.to_datetime(date).date())
dates
```

```python
data['Date'] = dates
data.tail(10)
```

```python
import datetime
base = data['Date'][0]
date_list = [base + datetime.timedelta(days=x) for x in range(2991)]
date_list[-1]
```

```python
data.index = data['Date']
```

```python
from numpy import NaN
data = data.reindex(date_list, fill_value=NaN)
data
```

```python
data = data.drop(['Name', 'Symbol'], axis=1)
data
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)
data = pd.DataFrame(imputer.fit_transform(data[['High', 'Low', 'Open', 'Close', 'Volume']]), columns = ['High', 'Low', 'Open', 'Close', 'Volume'])
data.index = dates
data
```
